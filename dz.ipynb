### 1–10: Простые функции
1. **Функция для приветствия гостей:**  
   Напишите функцию, которая принимает список имен гостей и выводит приветствие для каждого из них.
   
2. **Переводчик:**  
   Напишите функцию, которая принимает слово на одном языке и возвращает его перевод на другом языке. Можно использовать словарь для хранения переводов.

3. **Создание визиток:**  
   Напишите функцию, которая принимает имя, профессию и номер телефона, и возвращает текст визитки.

4. **Определение времени суток:**  
   Напишите функцию, которая принимает текущее время в формате HH:MM и возвращает, утро это, день, вечер или ночь.

5. **Функция описания погоды:**  
   Напишите функцию, которая принимает температуру и тип погоды (например, дождь или солнце) и возвращает подходящее описание дня.

6. **Симулятор чаевых:**  
   Напишите функцию, которая принимает сумму счета и процент чаевых, а затем возвращает сумму чаевых.

7. **Конструктор паролей:**  
   Напишите функцию, которая принимает длину пароля и возвращает случайный пароль из букв, цифр и символов.

8. **Именной калькулятор:**  
   Напишите функцию, которая принимает имя пользователя и строку арифметического выражения, а затем возвращает результат с упоминанием имени.

9. **Перевертыш слов:**  
   Напишите функцию, которая принимает строку и возвращает ее в обратном порядке.

10. **Шифратор:**  
    Напишите функцию, которая принимает строку и сдвигает каждую букву на указанное количество позиций в алфавите.

### 11–20: Работа с параметрами
11. **Уведомление о днях рождения:**  
    Напишите функцию, которая принимает имена и даты рождения, а затем возвращает список имен, у кого ближайший день рождения.

12. **Создание сообщений:**  
    Напишите функцию, которая принимает шаблон сообщения с заполнителями, например: `"Привет, {имя}! Твой заказ: {товар}"`, и параметры, чтобы заполнить их.

13. **Книга отзывов:**  
    Напишите функцию, которая принимает произвольное количество отзывов от пользователей и возвращает их в виде списка.

14. **Выбор случайного элемента:**  
    Напишите функцию, которая принимает список и возвращает случайный элемент из него.

15. **Определение палиндрома:**  
    Напишите функцию, которая проверяет, является ли введенное слово палиндромом.

16. **Генератор логинов:**  
    Напишите функцию, которая принимает имя и фамилию, а затем возвращает логин в формате `имя.фамилия`.

17. **Конвертер времени:**  
    Напишите функцию, которая принимает часы и минуты, а затем возвращает их в формате "столько-то секунд".

18. **Форматирование текста:**  
    Напишите функцию, которая принимает строку и возвращает ее с заглавной буквы, обрезанную до заданной длины.

19. **Произвольное приветствие:**  
    Напишите функцию, которая принимает имя и произвольное количество эмоций (например, “весело”, “радостно”) и возвращает приветствие с этими эмоциями.

20. **Фильтр слов:**  
    Напишите функцию, которая принимает список слов и букву, а затем возвращает только те слова, которые начинаются с этой буквы.

### 21–30: Сложные задачи
21. **Календарь задач:**  
    Напишите функцию, которая принимает список задач с их дедлайнами и возвращает задачи, дедлайн которых сегодня.

22. **Поиск книг:**  
    Напишите функцию, которая принимает список книг (название, автор, жанр) и возвращает только книги заданного жанра.

23. **Музыкальный плейлист:**  
    Напишите функцию, которая принимает список песен и их длительность, а затем возвращает общее время плейлиста.

24. **Сортировка пользователей:**  
    Напишите функцию, которая принимает список пользователей с их возрастом и возвращает их отсортированными по возрасту.

25. **Игра в слова:**  
    Напишите функцию, которая принимает слово и возвращает список всех возможных перестановок букв этого слова.

26. **Чат-бот:**  
    Напишите функцию, которая имитирует ответы на простые фразы, например: `"Как дела?" -> "Все отлично!"`.

27. **Генератор хэштегов:**  
    Напишите функцию, которая принимает список слов и возвращает строку, где каждое слово начинается с `#`.

28. **Анализ текста:**  
    Напишите функцию, которая принимает текст и возвращает количество слов, строк и самых часто встречающихся букв.

29. **Проверка формата email:**  
    Напишите функцию, которая проверяет, является ли переданная строка действительным email-адресом.

30. **Планировщик встреч:**  
    Напишите функцию, которая принимает список времени, когда вы заняты, и возвращает свободные интервалы для встреч.
